```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from time import sleep
```

```python
X, y = make_classification(n_samples=100,n_features=2, n_informative=2, n_redundant=0, n_classes=2, 
                              n_clusters_per_class=1, weights=None, random_state=1)
```

```python
y = [-1 if y_ else 1 for y_ in y]
```

```python
def plot(X, y, W):
    plt.scatter(X[:, 0], X[:, 1], c = y)
    
    x = np.arange(-1, 1, 0.2)
    y_val = [np.dot(W, [x_, 1]) for x_ in x]
    plt.plot(x, y_val, linewidth=2)
    plt.show()
```

```python
def scalar_multiply(scalar, vector):
    return np.array([scalar_*vector_ for scalar_, vector_ in zip(scalar, vector)])

def loss_function(X, y, W, C):
    return np.linalg.norm(W) + np.sum(C*np.maximum(0, 1 - np.multiply(y, np.dot(X, W))))

def derivative(X, y, W, C):
    indicator = np.array((np.maximum(0, 1 - np.multiply(y, np.dot(X, W))) == 0), dtype=np.int8)
    return W - C * np.sum(scalar_multiply(indicator, scalar_multiply(y, X)), axis=0)
```

```python
def main(epoch=20):
    W = np.array([1, 1])
    C = 10
    
    for i in range(epoch):
        loss = loss_function(X, y, W, C)
        if i%5==0:
            print(f"Epoch {i+1} loss {loss:.2f}, weight {np.round(W, 2)}")
            plot(X, y, W)
            
        W = W - 0.005*(1/X.shape[0])*derivative(X, y, W, C)
        sleep(0.5)
```

```python
main(300)
```
